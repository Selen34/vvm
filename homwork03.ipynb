- Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

len = x1(10,10,10) + x2(0,0,-10) = (10,10,0)

```python
def lvec(x: list):
    t = 0
    for i in x:
        t += i**2
    return t**0.5
```

## Задание 2

Почему прямые не кажутся перпендикулярными?

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from pylab import rcParams
```

```python
x = np.linspace(-5, 5, 21)
y1 = 3*x+1
y2 = (1/3)*x+1
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Ответ: гипотетически, из за разных масштабов осей

```python
x = np.linspace(-5, 5, 21)
rcParams["figure.figsize"] = (6,6)
y1 = 3*x + 1
y2 = (1/3)*x + 1
plt.axis('equal')
plt.grid()
plt.xlim((-5, 5))
plt.ylim((-5, 5))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 2. Пусть задана плоскость:

$A_1x + B_1y + C_1z + D_1 = 0$

и прямая:

\begin{equation}
\frac{x - x_1}{x_2 - x_1} = \frac{y-y_1}{y_2 - y_1} = \frac{z-z_1}{z_2 - z_1}
\end{equation}

Как узнать, принадлежит прямая плоскости или нет?

Ответ: подставить координаты прямой в уравнение плоскости

$A_1x_1 + B_1y_1 + C_1z_1 + D_1 = A_1x_2 + B_1y_2 + C_1z_2 + D_1$

Если равенство выполняется, значит прямая принадлежит плоскости

## Задание 9

### 1. Решите систему уравнений

\begin{equation}
\left\{ \begin{array}{ll}
y = x^2 - 1\\
e^x + x (1-y) = 1
\end{array} \right.
\end{equation}

\begin{equation}
\left\{ \begin{array}{ll}
y = x^2 - 1\\
y = 1 - \frac{1 - e^x}{x}
\end{array} \right.
\end{equation}

```python
def fy1(x):
    return x**2 - 1

def fy2(x):
    return 1 - (1 - np.exp(x))/x
```

```python
xx1 = np.linspace(-5, -0.01, 100)
xx2 = np.linspace(0.01, 5, 100)
plt.plot(xx1, fy1(xx1))
plt.plot(xx1, fy2(xx1))
plt.plot(xx2, fy1(xx2))
plt.plot(xx2, fy2(xx2))
plt.grid()
plt.show()
```

```python
from scipy.optimize import fsolve
```

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1, y-1+(1-np.exp(x))/x)
```

```python
x1, y1 = fsolve(equations, (-4, 5))
x2, y2 = fsolve(equations, (3, 3))
x3, y3 = fsolve(equations, (5, 5))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```

```python
xx1 = np.linspace(-5, -0.01, 100)
xx2 = np.linspace(0.01, 5, 100)
plt.plot(xx1, fy1(xx1))
plt.plot(xx1, fy2(xx1))
plt.plot(xx2, fy1(xx2))
plt.plot(xx2, fy2(xx2))
plt.scatter([x1, x2, x3],[y1, y2, y3])
plt.grid()
plt.show()
```
